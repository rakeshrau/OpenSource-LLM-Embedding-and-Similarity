## <font color='red'>Build the `sentence embedding` pipeline using ðŸ¤— Transformers Library<font>

```python
from sentence_transformers import SentenceTransformer
```

More info on [all-MiniLM-L6-v2](https://huggingface.co/sentence-transformers/all-MiniLM-L6-v2).

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
sentence1='Rakesh likes fruits'
```

```python
sentence2='Rakesh loves Mango'
```

```python
embeddings1 = model.encode(sentence1, convert_to_tensor=True)
embeddings2 = model.encode(sentence2, convert_to_tensor=True)
```

```python
print(embeddings1)
```

```python
print(embeddings2)
```

## <font color='red'>Find `similarity`<font>

```python
from sentence_transformers import util
```

```python
cosine_scores = util.cos_sim(embeddings1,embeddings2)
print(cosine_scores)
```
